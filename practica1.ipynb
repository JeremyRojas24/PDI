Ahora vamos a cargar nuestra imagen en escala de grises, pero la va a poner con las 3 capas rgb, sólo que como se supone que es en escala de grises, los valores se van a repetir.

```python
from skimage import io

image=io.imread('yin-yang.jpg')
io.imshow(image)
io.show()

print("Dimensiones de la imagen: ")
print(image.shape)
print("(filas (px), columnas(px), capas(rgb)")
print(image)

```

Cargando una imagen a color, ahora si va a tener un valor diferente en cada RGB.

```python
from skimage import io

image_color=io.imread('Spider-Man.jpg')
io.imshow(image_color)
io.show()
print(image_color.shape)
print(image_color)
```

Cargando una imagen a color con el parámetro de escala de grises, la va a generar en una matriz directamente con una sola capa, porque va a estar en escala de grises.

```python
from skimage import io

image_gris=io.imread('Spider-Man.jpg', as_gray=True)
io.imshow(image_gris)
io.show()
print(image_gris.shape)
print(image_gris)
```

Vamos a hacer la prueba de pasar a matriz de grises una imagen que ya está en escala de grises originalmente, pero evitando que se mapee en una matriz RGB.

```python
from skimage import io

image_doblegris=io.imread('yin-yang.jpg', as_gray=True)
io.imshow(image_doblegris)
io.show()

print("Dimensiones de la imagen: ")
print(image_doblegris.shape)
print("(filas (px), columnas(px), capas(rgb)")
print(image_doblegris)
```

Ahora vamos a sacarle el negativo a las imágenes, usando la fórmula s=(L-1)-r

```python
negativo = 255 - image
io.imshow(negativo)
io.show()
print(negativo.shape)
print(negativo)
```

Ejemplo pero ahora con la imagen a color

```python
negativo = 255 - image_color
io.imshow(negativo)
io.show()
print(negativo.shape)
print(negativo)
```

Por último vamos a probar con la imagen a color que se convierte a una matriz de escala de grises pura

```python
negativo = 255 - image_gris
io.imshow(negativo)
io.show()
print(negativo.shape)
print(negativo)
```

Vemos que en ese caso nos arroja una advertencia de que los valores de la matriz están fuera de rango por ser float, por lo que el programa nos está presentando la imagen con stretch contrast, es decir normalizada; pues se está "estirando" el rango de valores de intensidad para la imagen.

Veamos qué pasa con nuestra imagen que forzamos a ser una matriz de escala de grises.

```python
negativo = 255 - image_doblegris
io.imshow(negativo)
io.show()
print(negativo.shape)
print(negativo)
```

De igual manera se está normalizando y haciendo el negativo, por eso no arroja el mismo resultado que invertir los colores como tal. Y eso podemos apreciarlo haciendo la comparación de la LUT en nuestra imagen de escala de grises pero mapeada en la matriz RGB y esta de acá.
