Vamos a cargar la primer imagen en escala de gris y aplicarle la transformación logarítimica. 

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
   
image = cv2.imread('yin-yang.jpg') 
   
c = 255 / np.log(1 + np.max(image)) 
log_image = c * (np.log(image + 1)) 
   
log_image = np.array(log_image, dtype = np.uint8) 
   
plt.imshow(image) 
plt.show() 
plt.imshow(log_image) 
plt.show() 
```

Ahora vamos a probar con una imagen a color:

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
   
image = mpimg.imread('Spider-Man.jpg')
   
c = 255 / np.log(1 + np.max(image)) 
log_image = c * (np.log(image + 1)) 
   
log_image = np.array(log_image, dtype = np.uint8) 
   
imgplot = plt.imshow(image)
plt.show()
imgplot = plt.imshow(log_image)
plt.show()
```

Como segunda parte de la práctica, vamos a cargar a imagen en escala de gris y aplicarle la transformación de potencia.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

image = cv2.imread('yin-yang.jpg')
gamma = 4
gamma_corrected = np.array(255*(img / 255) ** gamma, dtype = 'uint8')
plt.imshow(image)
plt.show()
plt.imshow(gamma_corrected)
plt.show()
```

Ahora vamos a probar con una imagen a color:

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

image = mpimg.imread('Spider-Man.jpg')
gamma = 4
gamma_corrected = np.array(255*(image / 255) ** gamma, dtype = 'uint8')
imgplot = plt.imshow(image)
plt.show()
imgplot = plt.imshow(gamma_corrected)
plt.show()
```
