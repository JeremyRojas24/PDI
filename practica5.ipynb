Vamos a cargar las primeras imágenes en escala de gris y aplicarles una suma aritmética entre ellas.

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 

image1 = cv2.imread('yin-yang.jpg')
image2 = cv2.imread('vias.jpg')

suma = cv2.add(image1,image2)

plt.imshow(image1)
plt.show()
plt.imshow(image2)
plt.show()
plt.imshow(suma)
plt.show()
```

Ahora vamos a aplicarles una resta aritmética.

```python
resta = cv2.subtract(image1,image2)

plt.imshow(image1)
plt.show()
plt.imshow(image2)
plt.show()
plt.imshow(resta)
plt.show()
```

También podemos aplicar operaciones lógicas a las imágenes, vamos a empezar con un AND.

```python
img_and = cv2.bitwise_and(image1,image2)

plt.imshow(image1)
plt.show()
plt.imshow(image2)
plt.show()
plt.imshow(img_and)
plt.show()
```

Ahora vamos a aplicarles un OR.

```python
img_or = cv2.bitwise_or(image1,image2)

plt.imshow(image1)
plt.show()
plt.imshow(image2)
plt.show()
plt.imshow(img_or)
plt.show()
```

Y por último un XOR.

```python
img_xor = cv2.bitwise_xor(image1,image2)

plt.imshow(image1)
plt.show()
plt.imshow(image2)
plt.show()
plt.imshow(img_xor)
plt.show()
```
