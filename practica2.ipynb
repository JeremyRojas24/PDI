```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import numpy
import statistics
from scipy import stats as st

img = cv2.imread('Spider-Man.jpg')
print(img)

color = ('r','g','b')

for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
plt.imshow(img)
print ("La media es: ",numpy.mean(histr))

print("La mediana es:", statistics.median(histr))

moda=st.mode(histr)
print("La moda es:", moda)
varianza= histr.var()
print("La varianza es: ", varianza)
desviacion=histr.std()
print("La desviacion estandar es: ", desviacion)

```
